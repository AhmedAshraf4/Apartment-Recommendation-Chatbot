```python
import pandas as pd

# read csv
df = pd.read_csv("input.csv")

# map titles to prefixes
prefix_map = {
    "apartment": "AP",
    "studio": "AP",
    "penthouse": "AP",
    "duplex": "DU",
    "townhouse": "TH"
}

# create prefix column
df["prefix"] = df["title"].str.lower().map(prefix_map)

# create sequential numbers per prefix
df["num"] = df.groupby("prefix").cumcount() + 1

# format ID like AP001, DU002 etc
df["apartment_id"] = df["prefix"] + df["num"].astype(str).str.zfill(3)

# set the same email for all rows
df["agent_email"] = "ahmedashraf2452001@gmail.com"

# drop helper columns
df = df.drop(columns=["prefix", "num"])

# save updated csv
df.to_csv("output.csv", index=False)

print("Updated dataset saved to output.csv")
```
